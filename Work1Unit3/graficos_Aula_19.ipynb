```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import linspace
from scipy.stats import binom, poisson, hypergeom
```

## 1° GRÁFICO
    Distribuição binominal e distribuição de poisson

```python
outcomes_counts = linspace( 0, 10, 11)
# Distribuição Binomial
dist_binom = binom.pmf(outcomes_counts, 20, .05)
# Distribuição de Poisson
dist_poisson = poisson.pmf(outcomes_counts, 20*.05)

sns.set()
plt.plot(outcomes_counts, dist_binom, 'o', color='r')
plt.plot(outcomes_counts, dist_poisson, 'o', color='b')

plt.legend(loc='best',labels=['X~binomial(20,0.05)','Y~poisson(20*0.05)'])
plt.show()
```

## 2° GRÁFICO
    Distribuição hipergeométrica (pmf)

```python
outcomes_counts = linspace(0,20,21)
# Distribuição Hipergeométrica
dist_hipergeometrica = hypergeom.pmf(outcomes_counts,100,5,4)

sns.set()
plt.plot(outcomes_counts, dist_hipergeometrica, 'o', color='b')

plt.legend(loc='best',labels=['X~hypergeom(k,100,5,4)'])
plt.show()
```

## 3° GRÁFICO
    Distribuição de Hipergeometrica (cdf)

```python
outcomes_counts = linspace(0,20,21)
# Distribuição Hipergeométrica
dist_hipergeometrica = hypergeom.cdf(outcomes_counts,100,5,4)

sns.set()
plt.plot(outcomes_counts, dist_hipergeometrica, 'o', color='b')

plt.legend(loc='best',labels=['X~hypergeom(k,100,5,4)'])
plt.show()
```
