```python
# Questão 6
i = data[['status', 'enen-nota', 'a_ID']]
i.dropna()

n_ativo = i[i['status'] == 'ATIVO']
n_ativo= n_ativo.dropna()
n_ativo.replace('ATIVO', 1, inplace=True)

n_cancelado = i[i['status'] == 'CANCELADO']
n_cancelado = n_cancelado.dropna()
n_cancelado.replace('CANCELADO', 2, inplace=True)

n_trancado = i[i['status'] == 'TRANCADO']
n_trancado = n_trancado.dropna()
n_trancado.replace('TRANCADO', 3, inplace=True)

n_concluido = i[i['status'] == 'CONCLUIDO']
n_concluido = n_concluido.dropna()
n_concluido.replace('CONCLUIDO', 4, inplace=True)

n_formando = i[i['status'] == 'FORMANDO']
n_formando = n_formando.dropna() 
n_formando.replace('FORMANDO', 5, inplace=True)

n_formado = i[i['status'] == 'FORMADO']
n_formado = n_formado.dropna() 
n_formado.replace('FORMADO', 6, inplace=True)

fig = plt.figure(figsize=(12,8))
ax0 = fig.add_subplot(1,1,1)
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(1,1,1)
ax3 = fig.add_subplot(1,1,1)
ax4 = fig.add_subplot(1,1,1)
ax5 = fig.add_subplot(1,1,1)
ax0.scatter(n_ativo['a_ID'], n_ativo['enen-nota'], c='orange', label = 'ATIVO', alpha = 0.5)
ax1.scatter(n_cancelado['a_ID'], n_cancelado['enen-nota'], c= 'red', label = 'CANCELADO', alpha = 0.5)
ax2.scatter(n_trancado['a_ID'], n_trancado['enen-nota'], c= 'green', label = 'TRANCADO', alpha = 0.5)
ax3.scatter(n_concluido['a_ID'], n_concluido['enen-nota'], c= 'purple', label = 'CONCLUIDO', alpha = 0.5)
ax4.scatter(n_formando['a_ID'], n_formando['enen-nota'], c= 'blue', label = 'FORMANDO', alpha = 0.5)
ax5.scatter(n_formado['a_ID'], n_formado['enen-nota'], c= 'black', label = 'FORMADO', alpha = 0.5)
plt.legend(loc='upper left')

plt.title("Gráfico de dispersão dos Status e das notas do ENEM ")
plt.xlabel('Identificação do aluno')
plt.ylabel('Notas do Enem')
plt.show()
```
