```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import linspace
import math
import numpy as np
```

```python
#  ---------     MTTFs --------------

# Função que calcula o MTTF sem redundância
def MTTF_TMR(tam, laambda):
    m_t_t_f= tam*(1/laambda)   
    return m_t_t_f

# Função que calcula o MTTF sem redundância
def MTTF_2_paralela(tam, laambda):
    m_t_t_f= (pow(math.e,-2*laambda)-4*pow(math.e,-1*laambda)+3)/2*laambda  
    return m_t_t_f

# Função que calcula o MTTF sem redundância
def MTTF_3_paralela(tam, laambda):
    m_t_t_f= tam*(1/laambda)   
    return m_t_t_f

# Função que calcula o MTTF sem redundância
def MTTF_4_paralela(tam, laambda):
    m_t_t_f= tam*(1/laambda)   
    return m_t_t_f

# Função que calcula o MTTF sem redundância
def MTTF_sem_redundacia(tam, laambda):
    m_t_t_f= (1/laambda)   
    return m_t_t_f

def MTTF( list_lambdas ):
    list_lambdas = np.array(list_lambdas)
    return np.sum([1/i for i in list_lambdas])
# ---------- DISTRIBUIÇÕES DE CONFIABILIDADE  -------------

# Função que calcula a confiabilidade sem redundância no ponto t em relação a laambda
def sem_redundancia(laambda, t):
    i = pow(math.e,(-1)*laambda*t)
    return i
# Função que calcula a confiabilidade paralela, sobre 2 dispositivos, no ponto t em relação a laambda
def paralela_2(laambda,t):
    redundancia_paralela = (-1)*pow(math.e, -2*laambda*t)+2*pow(math.e,-1*laambda*t)
    return redundancia_paralela

# Função que calcula a confiabilidade paralela, sobre 3 dispositivos, no ponto t em relação a laambda
def paralela_3(laambda, t):
    redundancia_paralela = pow(math.e,-3*laambda*t) - 3*pow(math.e,-2*laambda*t) + 3*pow(math.e,-1*laambda*t)
    return redundancia_paralela

# Função que calcula a confiabilidade paralela, sobre 4 dispositivos, no ponto t em relação a laambda
def paralela_4(laambda,t):
    redundancia_paralela = (4*pow(math.e,-3*laambda*t))-(6*pow(math.e,-2*laambda*t)) -(pow(math.e, -4*laambda*t))+4*pow(math.e,-1*laambda*t)
    return redundancia_paralela

# Função que calcula a confiabilidade do NMR no ponto t em relação a laambda
def NMR(count_min, count_max, t, laambda):
    R_t = (count_min*pow(math.e,-1*count_max*laambda*t))-(count_max*pow(math.e,(-1)*count_min*laambda*t)) 
    return R_t
```

```python
outcomes_counts = linspace( 0, 1, 2000)
lista = [ 3, 5, 7, 9, 11 ]
for w in lista:
    tmr = [NMR(3,2,i,w) for i in outcomes_counts] # TMR
    mttf_tmr=[MTTF_TMR(2,w) for i in outcomes_counts]
    sem_reduncia = [sem_redundancia(w, i) for i in outcomes_counts] # Sem Redundância
    mttf_sem_redundancia=[MTTF_sem_redundacia(2,w) for i in outcomes_counts]
    CINCO_mr = [NMR(5,4,i,w) for i in outcomes_counts] # 5MR
    #mttf_5mr=[MTTF(4,w) for i in outcomes_counts]
    paralela_2_dispositvos = [paralela_2(i,w) for i in outcomes_counts] # Paralela com 2 dispositivos
    #mttf_paralela_2=[MTTF_2_paralela(paralela_2) for i in outcomes_counts]
    paralela_3_dispositvos = [paralela_3(w,i) for i in outcomes_counts] # Paralela com 3 dispositivos
    #mttf_paralela_3=MTTF(paralela_3)
    paralela_4_dispositvos = [paralela_4(w,i) for i in outcomes_counts] # Paralela com 4 dispositivos
    #mttf_paralela_4=MTTF(paralela_4)
    #m_t_t_f_sem_redundancia =[ MTTF_sem_redundacia(2, w) for i in outcomes_counts]
    #m_t_t_f_normal =[ MTTF_com_redundancia(2, w) for i in outcomes_counts]
   
    sns.set(font_scale=1.2,context='notebook', font='sans-serif', style='whitegrid')
    fig = plt.figure(figsize=(20,16))
    ax1= fig.add_subplot(2,2,++1)
 
    ax1.set_title('lambda = '+str(w))
    ax1.plot(mttf_sem_redundancia,outcomes_counts, linestyle='--', color='k',label= 'MTTF(Sem redundância)' )
    ax1.plot(mttf_tmr,outcomes_counts,         linestyle='--', label= 'MTTF(TMR)' )
    #ax1.plot(mttf_5mr,outcomes_counts,         linestyle='--', label= 'MTTF(5MR)' )
    #ax1.plot(mttf_paralela_2_dis ,outcomes_counts,         linestyle='--', label= 'MTTF(PARALELA 2)' )
    #ax1.plot(mttf_paralela_3,outcomes_counts,         linestyle='--', label= 'MTTF(PARALELA 3)' )
    #ax1.plot(mttf_paralela_4,outcomes_counts,         linestyle='--', label= 'MTTF(PARALELA 4)' )
    ax1.plot( outcomes_counts,paralela_2_dispositvos, color='m', label= 'Paralela (2 disp.)')
    ax1.plot( outcomes_counts,paralela_3_dispositvos, color='y', label= 'Paralela (3 disp.)')
    ax1.plot( outcomes_counts,paralela_4_dispositvos, color='b', label= 'Paralela (4 disp.)')
    ax1.plot( outcomes_counts,tmr,                    color='r', label= 'TMR')
    ax1.plot( outcomes_counts,sem_reduncia,           color='c', label= 'Sem redundância ')
    ax1.plot( outcomes_counts,CINCO_mr,            color='g', label= '5MR')
    ax1.set_ylabel("Reability(t)")
    ax1.set_xlabel("Time(t)")
    ax1.legend(loc='best', frameon=False)


    plt.show()
```
